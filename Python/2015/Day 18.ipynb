# Part 1

--- Day 18: Like a GIF For Your Yard ---

After the million lights incident, the fire code has gotten stricter: now, at most ten thousand lights are allowed. You arrange them in a 100x100 grid.

Never one to let you down, Santa again mails you instructions on the ideal lighting configuration. With so few lights, he says, you'll have to resort to animation.

Start by setting your lights to the included initial configuration (your puzzle input). A # means "on", and a . means "off".

Then, animate your grid in steps, where each step decides the next configuration based on the current one. Each light's next state (either on or off) depends on its current state and the current states of the eight lights adjacent to it (including diagonals). Lights on the edge of the grid might have fewer than eight neighbors; the missing ones always count as "off".

For example, in a simplified 6x6 grid, the light marked A has the neighbors numbered 1 through 8, and the light marked B, which is on an edge, only has the neighbors marked 1 through 5:

The state a light should have next is based on its current state (on or off) plus the number of neighbors that are on:

A light which is on stays on when 2 or 3 neighbors are on, and turns off otherwise.
A light which is off turns on if exactly 3 neighbors are on, and stays off otherwise.
All of the lights update simultaneously; they all consider the same current state before moving to the next.

After 4 steps, this example has four lights on.

In your grid of 100x100 lights, given your initial configuration, how many lights are on after 100 steps?



```python
data = aocd.get_data(day=18, year=2015)
data = data.splitlines()
# two grids: one current state -> second one updated state -> the copy second one to first one (because all update at once)
#two steps larger to have an empty border -> elems outside the border are always considered as 'off'
width, height = 102, 102
currentLightsGrid = [['.' for x in range(width)] for y in range(height)] 
tmpBoard = [['.' for x in range(width)] for y in range(height)] 

#initially fill grids
for i in range(len(data)):
    for j in range(len(data[i])):
        currentLightsGrid[i+1][j+1] = data[i][j]

# check neighbours

def checkTop(currentLightsGrid, x, y):
    if currentLightsGrid[x-1][y] == '#':
        return 1
    else:
        return 0

def checkBottom(currentLightsGrid, x, y):
    if currentLightsGrid[x+1][y] == '#':
        return 1
    else:
        return 0

def checkRight(currentLightsGrid, x, y):
    if currentLightsGrid[x][y+1] == '#':
        return 1
    else:
        return 0
    
def checkLeft(currentLightsGrid, x, y):
    if currentLightsGrid[x][y-1] == '#':
        return 1
    else:
        return 0 

def checkTopRight(currentLightsGrid, x, y):
    if currentLightsGrid[x-1][y+1] == '#':
        return 1
    else:
        return 0
    
def checkTopLeft(currentLightsGrid, x, y):
    if currentLightsGrid[x-1][y-1] == '#':
        return 1
    else:
        return 0 

def checkBottomRight(currentLightsGrid, x, y):
    if currentLightsGrid[x+1][y+1] == '#':
        return 1
    else:
        return 0
    
def checkBottomLeft(currentLightsGrid, x, y):
    if currentLightsGrid[x+1][y-1] == '#':
        return 1
    else:
        return 0 

#funtion to alter boards
def updateBoard(currentLightsGrid):
    #get placeholder board for every run
    tmpBoard = [['.' for x in range(width)] for y in range(height)] 
    
    #loop every position in the grid
    # off -> on if exactly 3 neighbours are on
    # on -> on if 2 or 3 are on, otherwise turn off
    # positions not on the board count as off
    # # == on ; . == off

    for x in range(1, len(currentLightsGrid)-1):
        for y in range(1, len(currentLightsGrid[x])-1):   
            
            #get neighbour positions and count activated lights
            onCounter = 0
            onCounter += checkTop(currentLightsGrid, x, y)
            onCounter += checkTopRight(currentLightsGrid, x, y)
            onCounter += checkTopLeft(currentLightsGrid, x, y)
            onCounter += checkRight(currentLightsGrid, x, y)
            onCounter += checkLeft(currentLightsGrid, x, y)
            onCounter += checkBottom(currentLightsGrid, x, y)
            onCounter += checkBottomRight(currentLightsGrid, x, y)
            onCounter += checkBottomLeft(currentLightsGrid, x, y)

            #alter light on tmp Board
            # light is on
            if(currentLightsGrid[x][y]) == '#':
                if onCounter == 2:
                    tmpBoard[x][y] = '#'
                elif onCounter == 3:
                    tmpBoard[x][y] = '#'
                else: 
                    tmpBoard[x][y] = '.'
            #light is off
            elif(currentLightsGrid[x][y]) == '.': 
                if onCounter == 3:
                    tmpBoard[x][y] = '#'  
                else: 
                    tmpBoard[x][y] = '.'         
        
    return tmpBoard

#number of board change cycles
cycles = 100
for i in range(cycles):
    
    currentLightsGrid = updateBoard(currentLightsGrid)

#count number of activated lights on the board
activatedLights = 0
for x in currentLightsGrid:
    for y in x:
        if y == '#':
            activatedLights += 1
        
print("Part 1: ", activatedLights)
```
